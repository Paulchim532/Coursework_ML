```python
### YOUR CODE HERE ###
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
```

```python
# параметры базовых алгоритмов
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)} # GradientBoostingClassifier

rfc_params = {'n_estimators': range(10, 100, 10), # RandomForestClassifier
              'min_samples_leaf': range(1, 5)}

svc_params = {'kernel': ['linear', 'rbf'], # SVC
              'C': np.arange(0.1, 1, 0.2)}

lr_params = {'C': np.arange(0.5, 1, 0.1)}

skf = StratifiedKFold(n_splits=5, random_state=17)
```

```python
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict
```

```python
from sklearn.metrics import accuracy_score



x_train, x_test, y_train, y_test = train_test_split(data_fm, 
                                                    data_labels,
                                                    train_size=0.3,
                                                    random_state=17)

gb = GradientBoostingClassifier(random_state=17)
gb_grid = GridSearchCV(gb, gbc_params, cv=skf)
gb_grid.fit(x_train, y_train)
print(gb_grid.best_params_)
gb_predicted = gb_grid.predict(x_test)
gb_accuracy = accuracy_score(y_test, gb_predicted)
print(f'GB Accuracy: {gb_accuracy}')

rf = RandomForestClassifier(random_state=17)
rf_grid = GridSearchCV(rf, rfc_params, cv=skf)
rf_grid.fit(x_train, y_train)
print(rf_grid.best_params_)
rf_predicted = rf_grid.predict(x_test)
rf_accuracy = accuracy_score(y_test, rf_predicted)
print(f'RF Accuracy: {rf_accuracy}')

svc = SVC(random_state=17, probability=True)
svc_grid = GridSearchCV(svc, svc_params, cv=skf)
svc_grid.fit(x_train, y_train)
print(svc_grid.best_params_)
svc_predicted = svc_grid.predict(x_test)
svc_accuracy = accuracy_score(y_test, svc_predicted)
print(f'SVC Accuracy: {svc_accuracy}')

lr = LogisticRegression(random_state=17)
lr_grid = GridSearchCV(lr, lr_params, cv=skf)
lr_grid.fit(x_train, y_train)
print(lr_grid.best_params_)
lr_predicted = lr_grid.predict(x_test)
lr_accuracy = accuracy_score(y_test, lr_predicted)
print(f'LR Accuracy: {lr_accuracy}')
```

# 4. Сравнение качества моделей 

```python
### YOUR CODE HERE ###

print(f'grad.boost. AUC: {roc_auc_score(y_test, gb_grid.predict_proba(x_test)[:, 1])}')

print(f'Rand.forest AUC: {roc_auc_score(y_test, rf_grid.predict_proba(x_test)[:, 1])}')

print(f'SVC AUC: {roc_auc_score(y_test, svc_grid.predict_proba(x_test)[:, 1])}')

print(f'Log.reg AUC: {roc_auc_score(y_test, lr_grid.predict_proba(x_test)[:, 1])}')
```
